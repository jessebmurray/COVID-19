```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('full-list-cumulative-total-tests-per-thousand-bar-chart.csv')
```

```python
df.head()
```

```python
s = df.groupby('Entity')['Total tests per thousand'].max().sort_values()
```

```python
top = s.shape[0]
plt.figure(figsize=(20, top*0.5))
plt.barh(list(s.to_dict().keys())[-top:], list(s.to_dict().values())[-top:])
```
