```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../files/condolences.csv', parse_dates=['Week'])
```

```python
df
```

```python
mask = pd.read_csv('mask.csv', parse_dates=['date'])
mask.date = mask.date.apply(lambda x: x.strftime('%Y-%m'))
# mask.set_index('date', inplace=True)
```

```python
short = pd.read_csv('shortness.csv', parse_dates=['date'])
short.date = short.date.apply(lambda x: x.strftime('%Y-%m'))
```

```python
push = pd.read_csv('pushups.csv', parse_dates=['date'])
push.date = push.date.apply(lambda x: x.strftime('%Y-%m'))
```

```python
df = df.rename(columns={'Week': 'date', 'condolences: (New York NY)': 'search'})
```

```python
df.set_index('date', inplace=True)
```

```python
dfm = df.resample("M").median() # gets the median for each month
```

```python
dfmc = df.copy()
```

```python
dfmc.date = dfmc.date.apply(lambda x: x.strftime('%Y-%m'))
```

```python
df.set_index('date', inplace=True)
dfm = df.resample("M") # gets the median for each month
```

```python
from matplotlib.ticker import PercentFormatter
plt.gca().yaxis.set_major_formatter(PercentFormatter(100))
sns.lineplot(x='date', y='search', data=dfmc, label='\'condolences\'')
# sns.lineplot(x='date', y='search', data=mask, label='\'mask\'')
sns.lineplot(x='date', y='search', data=short, label='\'shortness of breath\'')
# sns.lineplot(x='date', y='search', data=push, label='\'pushups\'')
plt.xticks(list(range(0, 61, 12)))
plt.xlabel('Date')
plt.ylabel('Percent of Max')
plt.yticks(range(0, 120, 20))
plt.legend()
plt.title('Google Searches in NYC')
plt.tight_layout()
# plt.savefig('covid1.png', dpi=300)
plt.show()
```
